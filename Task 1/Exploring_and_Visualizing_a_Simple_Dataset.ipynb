```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset
iris = sns.load_dataset('iris')
```

```python
print("Shape of dataset:", iris.shape)
print("\nColumns in dataset:", iris.columns.tolist())
print("\nFirst 5 rows of dataset:\n", iris.head())
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=iris, x='petal_length', kde=True, bins=20)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
features = iris.columns[:-1]  # excluding 'species'

plt.figure(figsize=(14, 8))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=iris, x='species', y=feature)
    plt.title(f'{feature.capitalize()} by Species')
plt.tight_layout()
plt.show()
```
